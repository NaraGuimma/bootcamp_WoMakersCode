Crie uma função que ao informar o sexo e a faixa etária de um suícidio é retornado as probilidades do método utilizado.


```python
suicidio_df['age.group'].unique()
```

```python
age_group_df = suicidio_df.groupby(['age.group', 'sex', 'method2'])['Freq'].sum()
age_group_df
```

```python
selecao = (suicidio_df['age.group'] == '55-65') & (suicidio_df['sex'] == 'female')
new_suicidio_df = suicidio_df[selecao].groupby('method2')['Freq'].sum().reset_index(name="soma_frequencia")
new_suicidio_df
```

```python
soma = new_suicidio_df['soma_frequencia'].sum()
soma
```

```python
new_suicidio_df['percentual'] = new_suicidio_df['soma_frequencia']/soma
new_suicidio_df
```

```python
def probability(sex, age_group):
  new_suicidio_df = pd.DataFrame()
  selecao = (suicidio_df['age.group'] == age_group) & (suicidio_df['sex'] == sex)
  new_suicidio_df = suicidio_df[selecao].groupby('method2')['Freq'].sum().reset_index(name="soma_frequencia")
  soma = new_suicidio_df['soma_frequencia'].sum()
  new_suicidio_df['probabilidade'] = new_suicidio_df['soma_frequencia']/soma
  new_suicidio_df.drop(columns = ['soma_frequencia'], inplace = True)
  new_suicidio_df.sort_values(by=['probabilidade'], ascending = False, inplace = True)
  return new_suicidio_df.round(4)
```

```python
probability('female', '55-65')
```

```python
probability('female', '10-20')
```
