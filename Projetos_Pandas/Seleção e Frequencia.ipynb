## Seleção e Frequencias

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep=";")
```

```python
dados
```

```python
#tipo = "Apartamento"

selecao = dados['Tipo'] == "Apartamento"
n1 = dados[selecao]
n1
```

```python
n1.shape[0]
```

```python
#tipo = "Casa", "Casa de Vila" ou "Casa de Condomínio"

selecao2 = dados['Tipo'].isin(["Casa", "Casa de Condomínio", "Casa de Vila"])
n2 = dados[selecao2]
n2
```

```python
selecao3 = (dados['Tipo'] == "Casa") | (dados['Tipo'] == "Casa de Condomínio") | (dados['Tipo'] == "Casa de Vila")
n3 = dados[selecao3]
n3
```

```python
# ter área entre 60 e 100 m²

selecao4 = (dados['Area'] >= 60) & (dados['Area'] <= 100)
n4 = dados[selecao4]
n4
```

```python
# mínimo 4 quartos e aluguel menos que R$ 2.000,00

selecao5 = (dados['Quartos'] >= 4) & (dados['Valor'] <= 2000.0)
n5 = dados[selecao5]
n5
```

```python
n5.shape[0]
```

```python
selecao6 = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000.0)
n6 = dados[selecao6]
n6
```
